```python
import os
import random
import pydub
import numpy as np
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
from pydub import AudioSegment
```

```python
path_output=' '
```

```python
class OverlapAudioChunck:

    def __init__(self,N_loop,OutputMainPath):

        self.InputMainPath=os.getcwd() # To get current work directory
        self.N_loop=N_loop
        self.N_files=5 # Number of files by defauld to rand
        self. I_files=1
        self.OutputMainPath=OutputMainPath

    def Get_Current_Path(self):
        return self.InputMainPath

    def List_All_Audios(self):
        Value_Files=[]
        Keys=[]
        if os.path.exists(self.Get_Current_Path()):
            for name in os.listdir(self.Get_Current_Path()):
                path_class = os.path.join(self.Get_Current_Path(), name)
                subFile=[]
                if os.path.isdir(path_class):
                    for audios in os.listdir(path_class):
                        complete_path = os.path.join(path_class, audios)
                        subFile.append(complete_path)
                    if len(subFile)==0:
                        continue
                    Keys.append(name)
                    Value_Files.append(subFile)
        return Keys,Value_Files

    def Pass_to_Dict(self):
        Key, Value=self.List_All_Audios()
        L=list()
        for i,j in zip(Key, Value):
            f={}
            f[i] = j
            L.append(f)
        L=np.array(L)
        return L

    def TakeOut_Audio(self,Obj):  #random of audios
        for key,value in Obj.items():
            rand=np.random.randint(len(value))
        return key,value[rand]

    def Read_Segment(self,segment):
        return pydub.AudioSegment.from_file(segment)

    def Overlab_Segment(self,Segment_list,path_to_import):
        L=list(map(self.Read_Segment, Segment_list))
        louder_segment_init=L[0]
        if len(L)!=1:
          for segment in L[1:]:
            louder_segment = segment
            Overlay_segment=louder_segment.overlay(louder_segment_init)
            louder_segment_init=Overlay_segment
            print('Done')
          louder_segment_init.export(path_to_import, format='wav')
          y, sr = librosa.load(path_to_import, mono = True, sr=32000)
          return y
        else:
            louder_segment_init.export(path_to_import, format='wav')
            y, sr = librosa.load(path_to_import, mono = True, sr=32000)
            print('Done one audio')
            return y

    def Write_Report(self,df,val1,val2,val3,val4,val5):
        df.loc[len(df)] = [val1,val2,val3,val4,val5]

    # Main Program
    def Shuffle_Segments(self,semilla):
        arr=self.Pass_to_Dict()  # array of dicts: key (name_file) and values (roots of audios, contents)
        
        df = pd.DataFrame(columns=['Iteration','Number Files', 'File','Segments Used','Segment_Name_Output'])
        Segment_Labels=[]
        List_Segment_Arrays=[]

        for N_species in range(1,self.N_files+1):
          for iterador in range(self.N_loop):
            np.random.seed(iterador+semilla) #Semilla
            np.random.shuffle(arr)
            name_to_output='Segment_'+str(((N_species-1)*self.N_loop)+1+iterador)+'.wav'
          
            path_output_file=path_output+name_to_output
            #path_output='/home/OverChunk-BirdClef2024/'+name_to_output
          
            Name_File=[]
            Root=[]
            List_to_OverLap=[]
            Only_name_audio=[]
              
            for k in arr[:N_species]:
                name_file,root=self.TakeOut_Audio(k)
                Name_File.append(name_file)
                List_to_OverLap.append(str(root))
                Root.append(os.path.basename(root))
                   
            print('==============> Loop',str(iterador+1))
            array=self.Overlab_Segment(List_to_OverLap,path_output_file) # calling Overlab function
            List_Segment_Arrays.append(array) ## List of arrays
            Segment_Labels.append(N_species)     ## List of labels
            self.Write_Report(df,((N_species-1)*self.N_loop)+1+iterador,N_species,Name_File,Root,name_to_output)

        Segment_Labels=np.array(Segment_Labels)
        Segment_Labels_copy = Segment_Labels.reshape((np.shape(Segment_Labels)[0],1))
        Name_report = 'Report.xlsx'
        df.to_excel(Name_report, index=False)

        Name_report = 'Report.csv'
        df.to_csv(Name_report, index=False)

        print(f"The data has been successfully exported '{Name_report}'")
        return Segment_Labels_copy,List_Segment_Arrays
```

```python
##### Execute  ####

# Create an object
Obj=OverlapAudioChunck(4000,path_output)
print('Object created')

# Call main function
M=Obj.Shuffle_Segments(42) #number seed
print('######### Matrix ################')
print(M)
print()
```
