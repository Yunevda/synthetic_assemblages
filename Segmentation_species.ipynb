```python
import librosa
from matplotlib import pyplot as plt
import numpy as np
import soundfile as sf
import os
```

```python
#The audio of each species should be divided into segments of five seconds in length.
path=" " #The path where the audio files are saved.
path_out = ' '

min_duration_seconds = 5
sr=32000
min_length_samples = min_duration_seconds * sr

for species in os.listdir(path):
    print(species)
    for audio in os.listdir(path+species):
        filename = path+species+'/'+audio
        y, sr = librosa.load(filename, mono = True, sr=sr)
        y = y/np.max(np.abs(y))
        
        if len(y) > min_length_samples: 
            rms = 0
            for i in range(0, len(y)-min_length_samples,1*sr):
                segment = y[i:i+5*sr]
                if np.sqrt(np.sum(segment**2)/len(segment))>rms:
                    clip = segment.copy()
                    rms = np.sqrt(np.sum(segment**2)/len(segment))
 
            if not os.path.exists(path_out+species):
                os.makedirs(path_out+species)

            filename_out = path_out+species+'/'+audio.split('.')[0]+'.wav'
            sf.write(filename_out, clip, sr, subtype='PCM_24') # Write out audio as 24bit PCM WAV
```
